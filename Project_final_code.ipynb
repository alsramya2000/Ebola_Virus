```python
import pandas as pd 
from datetime import datetime
import seaborn as sns
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
from sklearn.preprocessing import OneHotEncoder


# Convert the data into a transaction format

transaction_list = df['Symptoms'].str.split(', ').apply(set).tolist()


# Use TransactionEncoder to convert the data into a one-hot encoded format
te = TransactionEncoder()
te_ary = te.fit(transaction_list).transform(transaction_list)
one_hot_df = pd.DataFrame(te_ary, columns=te.columns_)

# Apply Apriori algorithm to find frequent itemsets with minimum support
min_support = 0.1
frequent_itemsets = apriori(one_hot_df, min_support=min_support, use_colnames=True)

# Print the frequent itemsets
print("Frequent Itemsets:")
print(frequent_itemsets)

# Apply Apriori algorithm to find frequent itemsets with minimum support
min_support = 0.04
frequent_itemsets = apriori(one_hot_df, min_support=min_support, use_colnames=True)

# Print the frequent itemsets
print("\nExpanded Frequent Itemsets:")
print(frequent_itemsets)
```

```python
# Apply Apriori algorithm to find frequent itemsets with minimum support of 0.05 (adjust as needed)
min_support = 0.05
frequent_itemsets = apriori(one_hot_df, min_support=min_support, use_colnames=True)

# Generate association rules from frequent itemsets with a minimum confidence of 0.3 (adjust as needed)
min_confidence = 0.3
association_rules_df = association_rules(frequent_itemsets, metric="confidence", min_threshold=min_confidence)

# Print the generated association rules
print("\nAssociation Rules:")
print(association_rules_df)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import KMeans

# Preprocess the Symptoms column by converting text into numerical features
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['Symptoms'])

# Apply K-Means clustering
n_clusters = 3  # You can adjust the number of clusters as needed
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)

# Print the cluster assignments
print(df[['Disease', 'Cluster']])
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Apply K-Means clustering for a range of k values
k_values = range(2, 11)
silhouette_scores = []
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    silhouette_scores.append(silhouette_score(X, kmeans.labels_))

# Plot the Silhouette Scores as a function of the number of clusters
plt.plot(k_values, silhouette_scores, 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score')
plt.show()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA


# Preprocess the Symptoms column by converting text into numerical features
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['Symptoms'])

# Apply K-Means clustering
n_clusters = 3  # You can adjust the number of clusters as needed
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
labels = kmeans.fit_predict(X)

# Add cluster labels to the DataFrame
df['Cluster'] = labels

# Analyzing cluster centers
cluster_centers = kmeans.cluster_centers_
cluster_centers_df = pd.DataFrame(cluster_centers, columns=vectorizer.get_feature_names_out())
print("Cluster Centers:")
print(cluster_centers_df)

# Evaluate the clustering performance using Silhouette Score
from sklearn.metrics import silhouette_score
silhouette_score_value = silhouette_score(X, labels)
print("Silhouette Score:", silhouette_score_value)

# Reduce the feature space to two dimensions using PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X.toarray())

# Add cluster labels to the DataFrame
df['Cluster'] = kmeans.labels_

# Plot the clusters
plt.figure(figsize=(8, 6))
for cluster_label in range(n_clusters):
    cluster_data = X_pca[df['Cluster'] == cluster_label]
    plt.scatter(cluster_data[:, 0], cluster_data[:, 1], label=f"Cluster {cluster_label}")
    
# Plot centroids
centroid_data = pca.transform(cluster_centers)
plt.scatter(centroid_data[:, 0], centroid_data[:, 1], marker='X', s=200, c='black', label='Centroids')

plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('K-Means Clustering of Symptoms and Diseases')
plt.legend()
plt.show()
```
